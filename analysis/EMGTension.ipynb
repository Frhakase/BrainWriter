```python
%matplotlib inline
import scipy
import numpy as np
from sklearn import svm
import copy
import itertools
from sklearn import preprocessing

#Spectrogram code used from: http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = np.zeros([fs/2,len(range(0, len(x)-framesamp, hopsamp) )])
    row_ctr = 0
    for i in range(0, len(x)-framesamp, hopsamp):
        weighted = w*x[i:i+framesamp]
        ps = np.abs(np.fft.fft(weighted))**2
        freqs = np.fft.fftfreq(w.size, 1/Fs)
        idxs = freqs>=0
        col = np.transpose(-10*log(ps[idxs]))
        X[:,row_ctr] = col
        row_ctr+=1
    return X

#Loops over the output signal and take the mode of the y at each time window
def block_output(y, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    Y = np.zeros(len(range(0, len(y)-framesamp, hopsamp)) )
    idx = 0
    for i in range(0, len(y)-framesamp, hopsamp):
        #What are the 1's and 0's at this time?
        if i+framesamp<len(y):
            vals = y[i:i+framesamp]    
        else:
            vals = y[i:]    
        #Mode returns [val, frequency]
        mode = scipy.stats.mode(vals)[0]
        if mode==0:
            mode=-1.0
        Y[idx] = mode 
        idx+=1
    return Y

def make_input_vector(x1_slice,x2_slice):
    NBINS_PER_FFT = 10
    bin_size = len(x1_slice)/NBINS_PER_FFT
    x1_bin = np.zeros(NBINS_PER_FFT)
    x2_bin = np.zeros(NBINS_PER_FFT)
    for i in range(0,NBINS_PER_FFT):
        x1_bin[i] = np.sum(x1_slice[i*bin_size:(i+1)*bin_size])
        x2_bin[i] = np.sum(x2_slice[i*bin_size:(i+1)*bin_size])
#    print len(x1_bin), len(x2_bin)
    results = np.concatenate([x1_bin, x2_bin])
    return results

    
#METADATA BOUT ABOUT THE CLINCHTEST FILE:
# Chan 2: motorcortex
# Chan 3: hairline (left side)

# 0:00-1:00: Settling the noise
# 1:00-3:15(ish): First experiment
# 3:15-5:30: talking and tweaking the experiments
# 5:30-7:30: Second experiment
# 7:30-8:00: rest
# 8:00-8:52: trying the rapid go/rest go/rest approach

# First experiment: Dan counts in, then counts "5...1" while conor imagines various arm motions (punching, freethrows, etc.)

# Second experiment: Dan is much more quiet. Just saying "Fire" and "Relax." Conor just thinks about clinching his right hand
    
CHAN1_IDX = 4 #motorcortex

data_file = "SavedData\OpenBCI-RAW-EMG_L_ARM_DAN1.txt"
###Create the spectrogram:
framesz = 1.0 # with a frame size of 50 milliseconds
hop = 1.0     # how far to go before defining the next frame  
Fs = 250.0


#FileLinks(data_directory) shows all the files
f = open(data_file,'r')
#Note that there was a glitch that switched the 'prompt' and 'timestamp' headers
lines = f.readlines()
signal = []

for line in lines:

    if line[0]=="%":
        continue
    
    elts = line.split(',')
    if len(elts)==1:
        continue
        
    signal.append(float(elts[CHAN1_IDX])) #I know apriori that's the right channel we care about

#convert the lists into numpy arrays    
signal = np.array(signal)

#Truncate the data to go in a specific time window
#Go for last minute: 8:00-8:52
# start_idx = length(signal)
# end_idx = -1

# signal1 = signal1[start_idx:]

print len(signal)

#de-mean the data
signal = signal - np.mean(signal)


#The two channels worth of signal are turned into spectrograms
x = signal
X = stft(x, Fs, framesz, hop);



#Convert the complex FFT results into usable magnitudes 
#NOTE: Should this be LOG?
X_spectrogram = scipy.absolute(X)


#Also, ignore the anything above ~50Hz or so (need to calculate exactly what it is)
X_spectrogram = X_spectrogram[:60,:]


fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.imshow(X_spectrogram, origin='lower', aspect='auto',interpolation='nearest');

plt.xlabel('Time');
plt.ylabel('Frequency');

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,4])
# plt.plot(X1_spectrogram[:,4])

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,14])
# plt.plot(X1_spectrogram[:,14])

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,-1])
# plt.plot(X1_spectrogram[:,-1])
```
