```python
%matplotlib inline
import scipy
import numpy as np
from sklearn import svm
import copy
import itertools
from sklearn import preprocessing

#Spectrogram code used from: http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = np.zeros([fs/2,len(range(0, len(x)-framesamp, hopsamp) )])
    row_ctr = 0
    for i in range(0, len(x)-framesamp, hopsamp):
        weighted = w*x[i:i+framesamp]
        ps = np.abs(np.fft.fft(weighted))**2
        freqs = np.fft.fftfreq(w.size, 1/Fs)
        idxs = freqs>=0
        col = np.transpose(-10*log(ps[idxs]))
        X[:,row_ctr] = col
        row_ctr+=1
    return X

#Loops over the output signal and take the mode of the y at each time window
def block_output(y, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    Y = np.zeros(len(range(0, len(y)-framesamp, hopsamp)) )
    idx = 0
    for i in range(0, len(y)-framesamp, hopsamp):
        #What are the 1's and 0's at this time?
        if i+framesamp<len(y):
            vals = y[i:i+framesamp]    
        else:
            vals = y[i:]    
        #Mode returns [val, frequency]
        mode = scipy.stats.mode(vals)[0]
        if mode==0:
            mode=-1.0
        Y[idx] = mode 
        idx+=1
    return Y

def remove_artifacts(x):
    mean = np.mean(x)
    std = np.std(x)
    
    for i in range(0, len(x)):
        if abs(x[i])>mean+3*std:
            x[i] = 0 #x[i-1]
    return x

def make_input_vector(x1_slice,x2_slice,x3_slice,x4_slice):
    bin_size = len(x1_slice)/NBINS_PER_FFT
    x1_bin = np.zeros(NBINS_PER_FFT)
    x2_bin = np.zeros(NBINS_PER_FFT)
    x3_bin = np.zeros(NBINS_PER_FFT)
    x4_bin = np.zeros(NBINS_PER_FFT)
    
    for i in range(0,NBINS_PER_FFT):
        x1_bin[i] = np.sum(x1_slice[i*bin_size:(i+1)*bin_size])
        x2_bin[i] = np.sum(x2_slice[i*bin_size:(i+1)*bin_size])
        x3_bin[i] = np.sum(x3_slice[i*bin_size:(i+1)*bin_size])
        x4_bin[i] = np.sum(x4_slice[i*bin_size:(i+1)*bin_size])
    results = np.concatenate([x1_bin, x2_bin,x3_bin,x4_bin])
    return results


NBINS_PER_FFT = 20 #for deciding the feature vector length

data_file = "outputfile_timestamp_1414982457.csv"

###Create the spectrogram:
framesz = 1.0 # 
hop = 1.0     # how far to go before defining the next frame  
Fs = 220.0


#Order of the files is
#Timestamp, expMode, Chan0,Chan1,Chan2,Chan3,
f = open(data_file,'r')

lines = f.readlines()
eeg = []
y = []
t = []

#timestamp,expMode,lOnset,rOnset,chan0,chan1,chan2,chan3
for line in lines:
    #print line
    #If it's the first line (headers,etc) ignor
    if line[0]=="prompt" or line[0]=="":
        continue
    
    elts = line.split(',')
    if len(elts)==1:
        continue
    if elts[0]=='timestamp': #if it's the header row
        continue
    t.append(float(elts[0]))
    y.append(float(elts[1]))
    eeg.append([float(elts[4]), float(elts[5]), float(elts[6]), float(elts[7])])

print "Total dataentries", len(y)
ave_period = np.mean(np.diff(t))
print "Average time between samples", ave_period
#convert the lists into numpy arrays    
eeg = np.array(eeg)
y = np.array(y)
t = np.array(t)

channel_means = np.mean(eeg,0)
for channel_idx in range(0,4):
    eeg[:,channel_idx] = eeg[:,channel_idx] - channel_means[channel_idx]

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.ylim([-1000,1000])
plt.plot(eeg);

# Make a sample waterfall plot
X0_spectrogram = scipy.absolute(stft(eeg[:,0], Fs, framesz, hop))
X1_spectrogram = scipy.absolute(stft(eeg[:,1], Fs, framesz, hop))
X2_spectrogram = scipy.absolute(stft(eeg[:,2], Fs, framesz, hop))
X3_spectrogram = scipy.absolute(stft(eeg[:,3], Fs, framesz, hop))

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.imshow(X1_spectrogram, origin='lower', aspect='auto',interpolation='nearest');

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.hold(True)
# plt.plot(X0_spectrogram[:,10]);
# plt.plot(X1_spectrogram[:,10]);
# plt.plot(X2_spectrogram[:,10]);
# plt.plot(X3_spectrogram[:,10]);
# plt.show()
```

```python
#Shrink each timeslice of the two spectrograms into a histogram format that can be used in 
x = [] #This is the final input signal into the FFT
for i in range(X1_spectrogram.shape[1]):
    input_vector = make_input_vector(X0_spectrogram[:,i], X1_spectrogram[:,i],X2_spectrogram[:,i],X3_spectrogram[:,i]) #,
    x.append(input_vector)

x_np = np.zeros([len(x),len(x[0])])
for row_idx in range(len(x)):
    x_np[row_idx,:] = x[row_idx]
x = x_np

#The output signal is extracted to math the inputs
atten = np.ones(len(y))
atten[y==0] = 0
a = block_output(atten, Fs, framesz, hop)

print "Len X={0} Y={1} spect={2}".format(len(x),len(a),X0_spectrogram.shape[1])
```
