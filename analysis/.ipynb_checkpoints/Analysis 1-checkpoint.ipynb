```python
#Spectrogram code used from: http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    
    #X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
    #                 for i in range(0, len(x)-framesamp, hopsamp)])
    X = np.zeros([fs/2,len(range(0, len(x)-framesamp, hopsamp) )])
    row_ctr = 0
    for i in range(0, len(x)-framesamp, hopsamp):
        weighted = w*x[i:i+framesamp]
        ps = np.abs(np.fft.fft(weighted))**2
        freqs = np.fft.fftfreq(w.size, 1/Fs)
        #idxs = np.argsort(freqs)
        idxs = freqs>=0
        col = np.transpose(-10*log(ps[idxs]))
        X[:,row_ctr] = col
        row_ctr+=1
    return X

#The built in time dimension is useless: it is discretized to whole seconds so we hardcode the Fs
Fs = 250.0
time = [(1/250.0)*x for x in range(0,time.size)]
channel_idxs = [0]#[3,5,7,9]
colors = ['b','r','g','k','-','c','m','y']

#The channels are marked starting from 1, but the csv starts with chan0
#So chan0=1N, chan1=2N, etc.
std_max = 0
for chan in channel_idxs:
    print d[chan,:20]
    chan_std = np.std(d[chan,:])
    if chan_std>std_max:
        std_max = chan_std

#plotting all the specifid channels
plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
for idx in range(0,len(channel_idxs)):
    plt.plot(time, d[channel_idxs[idx],:],colors[idx],label=headers[channel_idxs[idx]])

plt.ylim((-1.0* std_max,std_max))
print std_max
print channel_idxs


legend = plt.legend(loc='upper right', shadow=True)

plt.title("Total Time History for: " + data_file);


plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k');
for idx in range(0,len(channel_idxs)):
    channel = d[channel_idxs[idx],:]
    channel = channel - average(channel) #De-mean
    weighted = scipy.hamming(channel.size)*channel #Window
    ps = np.abs(np.fft.fft(weighted))**2 #PSD
    freqs = np.fft.fftfreq(weighted.size, 1/Fs) #Create set of frequency values for axis
    idxs = freqs>=0 #Get indices of all frequencies gte 0
    plt.plot(freqs[idxs], -10*log(ps[idxs]));#Plot using those frequenices
    
# plt.xlabel('Frequency');
# plt.ylabel('Log Magnitude of X^2');

# ###Create the spectrogram:
# framesz = 1.0 # with a frame size of 50 milliseconds
# hop = 1.0     # how far to go before defining the next frame  

# fig = plt.figure(num=None, figsize=(16,16), dpi=80, facecolor='w', edgecolor='k')
# for chan in range(0,len(channel_idxs)):
#     fig.add_subplot(len(channel_idxs)+1,1,chan+2) #make n+1 subplot, shift everything down
#     x = d[channel_idxs[chan],:]
#     X = stft(x, Fs, framesz, hop);

#     plt.imshow(scipy.absolute(X), origin='lower', aspect='auto',
#                      interpolation='nearest');
#     plt.xlabel('Time');
#     plt.ylabel('Frequency');
#     plt.title(headers[channel_idxs[chan]])

# #plot the prompt field 
# fig.add_subplot(len(channel_idxs)+1,1,1)
# plt.plot(time,d[1,:])
# plt.ylim(-.1,1.1)
# plt.xlim(0,X.shape[1])
# plt.show()
```

```python
#Generate sound
from scipy.io.wavfile import write
from math import pi, sin, log, exp

def sweep(f_start, f_end, interval, n_steps):
    b = log(f_end/f_start) / interval
    a = 2 * pi * f_start / b
    output = [0]*n_steps
    for i in range(n_steps):
        delta = i / float(n_steps)
        t = interval * delta
        g_t = a * exp(b * t)
        #print t, 3 * sin(g_t)
        output[i] = 3 * sin(g_t)
    return output

sweep_length = 15
data = sweep(5, 45, sweep_length, 44100*sweep_length)

#overwrite the data vector to prepend a synching signal like a clap
data = [min(data)]*(44100/3) + [max(data)]*(44100/3) + [0]*(44100/3) + data

#data = np.random.uniform(-1,1,44100) # 44100 random samples between -1 and 1

scaled = np.int16(data/np.max(np.abs(data)) * 32767)
write('test.wav', 44100, scaled)

plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(data)
```

```python
import scipy

#Spectrogram code used from: http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    
    #X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
    #                 for i in range(0, len(x)-framesamp, hopsamp)])
    X = np.zeros([fs/2,len(range(0, len(x)-framesamp, hopsamp) )])
    row_ctr = 0
    for i in range(0, len(x)-framesamp, hopsamp):
        weighted = w*x[i:i+framesamp]
        ps = np.abs(np.fft.fft(weighted))**2
        freqs = np.fft.fftfreq(w.size, 1/Fs)
        #idxs = np.argsort(freqs)
        idxs = freqs>=0
        col = np.transpose(-10*log(ps[idxs]))
        X[:,row_ctr] = col
        row_ctr+=1
    return X

def sweep(f_start, f_end, interval, n_steps):
    b = log(f_end/f_start) / interval
    a = 2 * pi * f_start / b
    output = [0]*n_steps
    for i in range(n_steps):
        delta = i / float(n_steps)
        t = interval * delta
        g_t = a * exp(b * t)
        #print t, 3 * sin(g_t)
        output[i] = 3 * sin(g_t)
    return output

sweep_length = 15
data = sweep(5, 45, sweep_length, 250*sweep_length)

#overwrite the data vector to prepend a synching signal like a clap
input_data = [min(data)]*(250/3) + [max(data)]*(250/3) + [0]*(250/3) + data

duration = len(input_data)

#Read in the sweep data
data_directory = '../Data021714 (sweep)/'

data_file = "SweepCorrectNoElectrode.txt"
#FileLinks(data_directory) shows all the files
f = open(data_directory + data_file,'r')
#Note that there was a glitch that switched the 'prompt' and 'timestamp' headers
lines = f.readlines()
signal = []
for line in lines:
    if line[0]=="%":
        continue
    elts = line.split(',')
    signal.append(float(elts[5])) #I know apriori that's the right channel we care about
    
signal = np.array(signal)
print "Length of signal: {0}".format(len(signal)/250.0)
print "Length of input: {0}".format(len(input_data))
start = np.where(signal>50)[0][0]
print start, len(signal)
signal = signal[start:]

###Create the spectrogram:
framesz = 1.0 # with a frame size of 50 milliseconds
hop = 1.0     # how far to go before defining the next frame  
Fs = 250.0
fig = plt.figure(num=None, figsize=(16,16), dpi=80, facecolor='w', edgecolor='k')
fig.add_subplot(3,1,1) # make n+1 subplot, shift everything down
x = signal
X = stft(x, Fs, framesz, hop);

plt.imshow(scipy.absolute(X), origin='lower', aspect='auto',interpolation='nearest');
plt.xlabel('Time');
plt.ylabel('Frequency');

#plot the prompt field 
fig.add_subplot(3,1,2)
plt.plot(signal)
plt.show()

fig.add_subplot(3,1,3)
plt.plot(input_data)
plt.show()


#Before we run the frf, recall that there is a "clap" in the input that we need to skip
offset = 250
input_data = input_data[offset:]
Finput = np.fft.fft(input_data)
#print len(input_data)
Foutput = np.fft.fft(signal[offset:offset+len(input_data)])

FRF = Foutput/Finput
PS = np.abs(FRF)
delta_f = Fs/len(PS)
print delta_f
fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(PS[len(PS)/2:len(PS)/2+500])
```
