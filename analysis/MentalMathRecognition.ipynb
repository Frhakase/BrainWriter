```python
%matplotlib inline
import scipy
import numpy as np
from sklearn import svm
import copy
import itertools
from sklearn import preprocessing

#Spectrogram code used from: http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = np.zeros([fs/2,len(range(0, len(x)-framesamp, hopsamp) )])
    row_ctr = 0
    for i in range(0, len(x)-framesamp, hopsamp):
        weighted = w*x[i:i+framesamp]
        ps = np.abs(np.fft.fft(weighted))**2
        freqs = np.fft.fftfreq(w.size, 1/Fs)
        idxs = freqs>=0
        col = np.transpose(-10*log(ps[idxs]))
        X[:,row_ctr] = col
        row_ctr+=1
    return X

#Loops over the output signal and take the mode of the y at each time window
def block_output(y, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    Y = np.zeros(len(range(0, len(y)-framesamp, hopsamp)) )
    idx = 0
    for i in range(0, len(y)-framesamp, hopsamp):
        #What are the 1's and 0's at this time?
        if i+framesamp<len(y):
            vals = y[i:i+framesamp]    
        else:
            vals = y[i:]    
        #Mode returns [val, frequency]
        mode = scipy.stats.mode(vals)[0]
        if mode==0:
            mode=-1.0
        Y[idx] = mode 
        idx+=1
    return Y

def remove_artifacts(x):
    mean = np.mean(x)
    std = np.std(x)
    
    for i in range(0, len(x)):
        if abs(x[i])>mean+3*std:
            x[i] = 0 #x[i-1]
    return x

def make_input_vector(x1_slice,x2_slice):
    bin_size = len(x1_slice)/NBINS_PER_FFT
    x1_bin = np.zeros(NBINS_PER_FFT)
    x2_bin = np.zeros(NBINS_PER_FFT)
    for i in range(0,NBINS_PER_FFT):
        x1_bin[i] = np.sum(x1_slice[i*bin_size:(i+1)*bin_size])
        x2_bin[i] = np.sum(x2_slice[i*bin_size:(i+1)*bin_size])
    results = np.concatenate([x1_bin, x2_bin])
    return results

# def make_input_vector(x1_slice,x2_slice, x3_slice):
#     bin_size = len(x1_slice)/NBINS_PER_FFT
#     x1_bin = np.zeros(NBINS_PER_FFT)
#     x2_bin = np.zeros(NBINS_PER_FFT)
#     x3_bin = np.zeros(NBINS_PER_FFT)
#     for i in range(0,NBINS_PER_FFT):
#         x1_bin[i] = np.sum(x1_slice[i*bin_size:(i+1)*bin_size])
#         x2_bin[i] = np.sum(x2_slice[i*bin_size:(i+1)*bin_size])
#         x3_bin[i] = np.sum(x3_slice[i*bin_size:(i+1)*bin_size])
#     results = np.concatenate([x1_bin, x2_bin,x3_bin])
#     return results

    
#METADATA BOUT ABOUT THE CLINCHTEST FILE:

# Chan 1: Left forehead lobe
# Chan 2: Middle ish of forehead
# Chan 3: Right lobe on forhead
# Reference on behidn left ear
# Using the cheap gold sensors and vaseline-ish stuff

# Ran the BetaMaker program in OpenFrameworks for a few minutes, not sure how  long
    

CHAN1_IDX = 2
CHAN2_IDX = 3 
CHAN3_IDX = 4 
OUTPUT_IDX = 1
NBINS_PER_FFT = 10 #for deciding the feature vector length

data_file = "betamakerwindows.csv"#"mentalmath052414.csv"
###Create the spectrogram:
framesz = 1.0 # with a frame size of 50 milliseconds
hop = 1.0     # how far to go before defining the next frame  
Fs = 250.0


#FileLinks(data_directory) shows all the files
f = open(data_file,'r')
#Note that there was a glitch that switched the 'prompt' and 'timestamp' headers
lines = f.readlines()
signal1 = []
signal2 = []
signal3 = []
firesignal = []
for line in lines:
    #print line
    #If it's the first line (headers,etc) ignor
    if line[0]=="prompt" or line[0]=="":
        continue
    
    elts = line.split(',')
    if len(elts)==1:
        continue
    if elts[0]=='timestamp':
        continue
        
    signal1.append(float(elts[CHAN1_IDX])) #I know apriori that's the right channel we care about
    signal2.append(float(elts[CHAN2_IDX])) #I know apriori that's the right channel we care about
    signal3.append(float(elts[CHAN3_IDX])) #I know apriori that's the right channel we care about
    firesignal.append(float(elts[OUTPUT_IDX])) #The "Fire!" signal is appended at the bottom

print len(firesignal)

#convert the lists into numpy arrays    
signal1 = np.array(signal1)
signal2 = np.array(signal2)
signal3 = np.array(signal2)
firesignal = np.array(firesignal)


#remove the artifacts
signal1 = remove_artifacts(signal1)
signal2 = remove_artifacts(signal2)
signal3 = remove_artifacts(signal3)

#de-mean the data
signal1 = signal1 - np.mean(signal1)
signal2 = signal2 - np.mean(signal2)
signal2 = signal3 - np.mean(signal3)

#The two channels worth of signal are turned into spectrograms
x1 = signal1
X1 = stft(x1, Fs, framesz, hop);
x2 = signal2
X2 = stft(x2, Fs, framesz, hop);
#x3 = signal3
#X3 = stft(x3, Fs, framesz, hop);

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.ylim([-1000,1000])
plt.plot(x1);

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x2);


#The output signal is extracted to math the inputs
y = block_output(firesignal, Fs, framesz, hop);

#Convert the complex FFT results into usable magnitudes 
#NOTE: Should this be LOG?
X1_spectrogram = scipy.absolute(X1)
X2_spectrogram = scipy.absolute(X2)
#X3_spectrogram = scipy.absolute(X3)

#Also, ignore the anything above ~50Hz or so (need to calculate exactly what it is)
X1_spectrogram = X1_spectrogram[:60,:]
X2_spectrogram = X2_spectrogram[:60,:]
#X3_spectrogram = X3_spectrogram[:60,:]

#Shrink each timeslice of the two spectrograms into a histogram format that can be used in 
x = [] #This is the final input signal into the FFT
for i in range(X1_spectrogram.shape[1]):
    input_vector = make_input_vector(X1_spectrogram[:,i], X2_spectrogram[:,i]) #,X3_spectrogram[:,i]
    x.append(input_vector)

x_np = np.zeros([len(x),len(x[0])])
for row_idx in range(len(x)):
    x_np[row_idx,:] = x[row_idx]
x = x_np

print "Length of input vector: " , len(input_vector)
#NOW we have data ready for SVM!
#Do a leave-one-out analysis to see how it performs
number_corrects = 0
confusion_matrix = np.zeros([2,2])

#fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
#plt.imshow(x)
#print x.shape
for loo_idx in range(0,len(x)):
    #Create the test and train data 
    y_test = y[loo_idx]
    x_test = x[loo_idx,:]

    y_train = np.array(list(itertools.compress(y, [i != loo_idx for i in range(len(y))])))
    x_train = np.delete(x,loo_idx,axis=0)

    #Debugging
    if np.isinf(x_train).any():
        print "xtrain", loo_idx
        break

    #Preprocess using the awesome preprocessing pipeline
    #documentation here: http://scikit-learn.org/stable/modules/preprocessing.html#preprocessing
    scaler = preprocessing.StandardScaler().fit(x_train)
    x_train_processed = scaler.transform(x_train)                               
    
    clf = svm.SVC(C=10,class_weight='auto')
    #Trying to do a L1 regularization
    #clf = svm.LinearSVC(C=1,class_weight='auto', loss='l2', penalty='l1', dual=False)
    clf.fit(x_train_processed, y_train)
        
    x_test_processed= scaler.transform(x_test)
    
    guess = clf.predict(x_test_processed)
    
    #print y_test, guess
    
    #Populate the confusion matrix
    guess_idx = int(guess[0])
    if guess_idx == -1:
        guess_idx = 0
    actual_idx = y_test
    if actual_idx == -1:
        actual_idx = 0    
    confusion_matrix[actual_idx, guess_idx] +=1
    
    if int(y_test) == int(guess[0]):
        number_corrects +=1
    
    #break 

print "Total accuracy: {0}/{1} = {2}".format(number_corrects, len(x), float(number_corrects)/len(x))
print confusion_matrix

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
fig.add_subplot(2,1,1) 
plt.imshow(X1_spectrogram[:60,:], origin='lower', aspect='auto',interpolation='nearest');
fig.add_subplot(2,1,2) 
plt.imshow(X2_spectrogram[:60,:], origin='lower', aspect='auto',interpolation='nearest');

plt.xlabel('Time');
plt.ylabel('Frequency');

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,4])
# plt.plot(X1_spectrogram[:,4])

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,14])
# plt.plot(X1_spectrogram[:,14])

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,-1])
# plt.plot(X1_spectrogram[:,-1])
```

```python
#Just plotting the data to make sure it's reasonable from OpenFrameworks
% matplotlib inline
import scipy
import numpy as np
from sklearn import svm
import copy
import itertools
from sklearn import preprocessing
from scipy.signal import butter, lfilter 

#Spectrogram code used from: http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = np.zeros([fs/2,len(range(0, len(x)-framesamp, hopsamp) )])
    row_ctr = 0
    for i in range(0, len(x)-framesamp, hopsamp):
        weighted = w*x[i:i+framesamp]
        ps = np.abs(np.fft.fft(weighted))**2
        freqs = np.fft.fftfreq(w.size, 1/Fs)
        idxs = freqs>=0
        col = np.transpose(-10*log(ps[idxs]))
        X[:,row_ctr] = col
        row_ctr+=1
    return X

def make_input_vector(x1_slice,x2_slice):
    bin_size = len(x1_slice)/NBINS_PER_FFT
    x1_bin = np.zeros(NBINS_PER_FFT)
    x2_bin = np.zeros(NBINS_PER_FFT)
    for i in range(0,NBINS_PER_FFT):
        x1_bin[i] = np.sum(x1_slice[i*bin_size:(i+1)*bin_size])
        x2_bin[i] = np.sum(x2_slice[i*bin_size:(i+1)*bin_size])
    results = np.concatenate([x1_bin, x2_bin])
    return results

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a
 
def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

def readDataFile(data_file, input_channel_indices, outputchannel_idx):
    f = open(data_file,'r')
    #Note that there was a glitch that switched the 'prompt' and 'timestamp' headers
    lines = f.readlines()
    inputsignals = []
    labelsignal = []
    for line in lines:
        #If it's the first line (headers,etc) ignore
        if line[0]=="prompt" or line[0]=="":
            continue
        
        elts = line.split(',')
        if len(elts)==1:
            continue
        if elts[0]=='timestamp':
            continue
        
        row = []
        for in_idx in input_channel_indices:
            row.append(float(elts[in_idx]))
            
        inputsignals.append(row)
        
        if outputchannel_idx:
            labelsignal.append(float(elts[outputchannel_idx])) #The "Fire!" signal is appended at the bottom    

        f.close()
    
    count = len(inputsignals)
    input_data = np.zeros([count, len(input_channel_indices)])
    for i in range(0,count):
        for j in range(0,len(input_channel_indices)):
            input_data[i,j] = inputsignals[i][j]
    
    return (input_data, labelsignal)

def remove_artifacts(x):
    mean = np.mean(x)
    std = np.std(x)
    
    for i in range(1, len(x)):
        if abs(x[i])>mean+3*std:
            x[i] = x[i-1]  
    return x

#Loops over the output signal and take the mode of the y at each time window
def block_output(y, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    Y = np.zeros(len(range(0, len(y)-framesamp, hopsamp)) )
    idx = 0
    for i in range(0, len(y)-framesamp, hopsamp):
        #What are the 1's and 0's at this time?
        if i+framesamp<len(y):
            vals = y[i:i+framesamp]    
        else:
            vals = y[i:]    
        #Mode returns [val, frequency]
        mode = scipy.stats.mode(vals)[0]
        if mode==0:
            mode=-1.0
        Y[idx] = mode 
        idx+=1
    return Y

NBINS_PER_FFT = 10 #for deciding the feature vector length

data_file = "/Users/dangoodwin/Code/OpenFrameworks/apps/Brainwriter/Data060914/firstmentalmath(post1drink).csv"

###Create the spectrogram:
framesz = 1.0 # with a frame size of 50 milliseconds
hop = 1.0     # how far to go before defining the next frame  
Fs = 250.0
lowcut_freq = 2.
highcut_freq = 70.

(inputdata, outputdata) = readDataFile(data_file, [0,1],2) 


#convert the lists into numpy arrays    
signal1 = np.array(inputdata[:,0])
signal2 = np.array(inputdata[:,1])
firesignal = np.array(outputdata)

#remove the artifacts
signal1 = remove_artifacts(signal1)
signal2 = remove_artifacts(signal2)

#bandpass the data
signal1 = butter_bandpass_filter(signal1, lowcut_freq, highcut_freq, Fs, order=6)
signal2 = butter_bandpass_filter(signal1, lowcut_freq, highcut_freq, Fs, order=6)

#truncate the data to deal with the edge effects of the bandpass filter
STARTIDX=1000
signal1 = signal1[STARTIDX:]
signal2 = signal2[STARTIDX:]
firesignal = firesignal[STARTIDX:]

#The two channels worth of signal are turned into spectrograms
x1 = signal1
X1 = stft(x1, Fs, framesz, hop);
X1_spectrogram = scipy.absolute(X1)

x2 = signal2
X2 = stft(x2, Fs, framesz, hop);
X2_spectrogram = scipy.absolute(X2)

#The output signal is extracted to math the inputs
y = block_output(firesignal, Fs, framesz, hop);

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x1,'r');
plt.plot(x2,'b');


fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.imshow(X1_spectrogram[:60,:], origin='lower', aspect='auto',interpolation='nearest');

plt.xlabel('Time');
plt.ylabel('Frequency');


#Shrink each timeslice of the two spectrograms into a histogram format that can be used in 
x = [] #This is the final input signal into the FFT
for i in range(X1_spectrogram.shape[1]):
    input_vector = make_input_vector(X1_spectrogram[:,i], X2_spectrogram[:,i]) #,X3_spectrogram[:,i]
    x.append(input_vector)

x_np = np.zeros([len(x),len(x[0])])
for row_idx in range(len(x)):
    x_np[row_idx,:] = x[row_idx]
x = x_np

print "Length of input vector: " , len(input_vector)
#NOW we have data ready for SVM!
#Do a leave-one-out analysis to see how it performs
number_corrects = 0
confusion_matrix = np.zeros([2,2])

#fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
#plt.imshow(x)
#print x.shape
for loo_idx in range(0,len(x)):
    #Create the test and train data 
    y_test = y[loo_idx]
    x_test = x[loo_idx,:]

    y_train = np.array(list(itertools.compress(y, [i != loo_idx for i in range(len(y))])))
    x_train = np.delete(x,loo_idx,axis=0)

    #Debugging
    if np.isinf(x_train).any():
        print "xtrain", loo_idx
        break

    #Preprocess using the awesome preprocessing pipeline
    #documentation here: http://scikit-learn.org/stable/modules/preprocessing.html#preprocessing
    scaler = preprocessing.StandardScaler().fit(x_train)
    x_train_processed = scaler.transform(x_train)                               
    
    clf = svm.SVC(C=10,class_weight='auto')
    #Trying to do a L1 regularization
    #clf = svm.LinearSVC(C=1,class_weight='auto', loss='l2', penalty='l1', dual=False)
    clf.fit(x_train_processed, y_train)
        
    x_test_processed= scaler.transform(x_test)
    
    guess = clf.predict(x_test_processed)
    
    #print y_test, guess
    
    #Populate the confusion matrix
    guess_idx = int(guess[0])
    if guess_idx == -1:
        guess_idx = 0
    actual_idx = y_test
    if actual_idx == -1:
        actual_idx = 0    
    confusion_matrix[actual_idx, guess_idx] +=1
    
    if int(y_test) == int(guess[0]):
        number_corrects +=1
    
    #break 

print "Total accuracy: {0}/{1} = {2}".format(number_corrects, len(x), float(number_corrects)/len(x))
print confusion_matrix

fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(X2_spectrogram[:,4])
plt.plot(X1_spectrogram[:,4])

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,14])
# plt.plot(X1_spectrogram[:,14])

# fig = plt.figure(num=None, figsize=(16,8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(X2_spectrogram[:,-1])
# plt.plot(X1_spectrogram[:,-1])
```
